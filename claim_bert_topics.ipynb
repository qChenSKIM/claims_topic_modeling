# Parameters in BertTopic:
- language: 
  * all-MiniLM-L6-v2: english
  * paraphrase-multilingual-MiniLM-L12-v2: multilingual
- top_n_words: The number of words per topic to extract (<10)
- n_gram_range: 1-3 (3)
- embedding_model:
  * SentenceTransformers
  * Flair
  * Spacy
  * Gensim
  * USE (TF-Hub)
  * https://www.sbert.net/docs/pretrained_models.html
- umap_model:
    ```
    self.umap_model = umap_model or UMAP(n_neighbors=15,
    n_components=5,
    min_dist=0.0,
    metric='cosine',
    low_memory=self.low_memory)
    ```
- hdbscan_model:  pass in any clustering algorithm as long as it has`.fit` and `.predict` functions along with the `.labels_` variable

## Setting3:

- hdbscan:
  * `min_cluster_size`: Ideally this is a relatively intuitive parameter to select – set it to the smallest size grouping that you wish to consider a cluster.

  * The larger the value of `min_samples` you provide, the more conservative the clustering – more points will be declared as noise, and clusters will be restricted to progressively more dense areas. 


````python
import hdbscan
hdbModel = hdbscan.HDBSCAN(min_cluster_size=3, min_samples=1)

from sentence_transformers import SentenceTransformer
emb_model = SentenceTransformer('all-mpnet-base-v2')


topic_model = BERTopic(language = "english",embedding_model = emb_model,hdbscan_model=hdbModel)
topics, probs = topic_model.fit_transform(docs)
````
